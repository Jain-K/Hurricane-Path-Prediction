```python
# Section 2: Define Utility Functions
def normalize_score(series):
    """Normalize values to 0-100 scale"""
    min_val = series.min()
    max_val = series.max()
    return ((series - min_val) / (max_val - min_val)) * 100

def calculate_population_score(df):
    """Calculate population vulnerability score"""
    weights = {
        'young_old_weight': 0.6,
        'total_pop_weight': 0.4
    }

    return normalize_score(
        (df['Total_Young'] + df['Total_Old']) * weights['young_old_weight'] +
        df['Total_Total population'] * weights['total_pop_weight']
    )

def calculate_economic_score(df):
    """Calculate economic vulnerability score"""
    weights = {
        'poor': 0.4,
        'lower_middle': 0.3,
        'middle': 0.2,
        'high_income': 0.05,
        'wealthy': 0.05
    }

    return normalize_score(
        df['Poor'] * weights['poor'] +
        df['Lower_Middle'] * weights['lower_middle'] +
        df['Middle'] * weights['middle'] +
        df['High_Income'] * weights['high_income'] +
        df['Wealthy'] * weights['wealthy']
    )

def calculate_mobility_score(df):
    """Calculate mobility vulnerability score"""
    weights = {
        'no_vehicle': 0.4,
        'one_vehicle': 0.3,
        'two_vehicle': 0.2,
        'three_vehicle': 0.1
    }

    return normalize_score(
        (df['Households_Total'] - df['1 Vehicle Households']) * weights['no_vehicle'] +
        df['1 Vehicle Households'] * weights['one_vehicle'] +
        df['2 Vehicle Households'] * weights['two_vehicle'] +
        df['3 Vehicle Households'] * weights['three_vehicle']
    )
```

```python
# Section 3: Main Analysis Functions
def process_single_day(df):
    """Process and score a single day's data"""
    results = df.copy()

    # Calculate component scores
    results['population_score'] = calculate_population_score(results)
    results['economic_score'] = calculate_economic_score(results)
    results['mobility_score'] = calculate_mobility_score(results)

    # Calculate total impact score
    weights = {'population': 0.4, 'economic': 0.3, 'mobility': 0.3}
    results['total_impact_score'] = (
        results['population_score'] * weights['population'] +
        results['economic_score'] * weights['economic'] +
        results['mobility_score'] * weights['mobility']
    )

    return results

def analyze_all_days(all_dfs):
    """Process all days and compile results"""
    # Process each day
    processed_dfs = []
    for df in all_dfs:
        day_results = process_single_day(df)
        processed_dfs.append(day_results)

    # Combine all results
    combined_results = pd.concat(processed_dfs, keys=[df['Day'].iloc[0] for df in all_dfs])
    return processed_dfs, combined_results
```

```python
# Section 4: Analysis and Insights Functions
def identify_most_affected_areas(combined_results):
    """Identify most affected counties and states"""
    # County level analysis
    county_summary = combined_results.groupby(['County', 'State']).agg({
        'total_impact_score': ['mean', 'max'],
        'population_score': 'mean',
        'economic_score': 'mean',
        'mobility_score': 'mean',
        'Total_Total population': 'first'
    }).round(2)

    # State level analysis
    state_summary = combined_results.groupby('State').agg({
        'total_impact_score': ['mean', 'max'],
        'Total_Total population': 'sum'
    }).round(2)

    # Get top affected areas
    most_affected_counties = county_summary.nlargest(10, ('total_impact_score', 'mean'))
    most_affected_states = state_summary.nlargest(5, ('total_impact_score', 'mean'))

    return most_affected_counties, most_affected_states
```

```python
# Section 5: Visualization Functions (Updated)

def plot_impact_trends(processed_dfs):
    """Plot impact score trends over days"""
    plt.figure(figsize=(12, 6))

    days = [df['Day'].iloc[0] for df in processed_dfs]
    avg_scores = [df['total_impact_score'].mean() for df in processed_dfs]
    max_scores = [df['total_impact_score'].max() for df in processed_dfs]

    plt.plot(days, avg_scores, marker='o', label='Average Impact Score')
    plt.plot(days, max_scores, marker='s', label='Maximum Impact Score')

    plt.title('Hurricane Impact Scores Over Time')
    plt.xlabel('Day')
    plt.ylabel('Impact Score')
    plt.legend()
    plt.grid(True)

    return plt.gcf()

def plot_vulnerability_heatmap(most_affected_counties):
    """Create heatmap of vulnerability scores for most affected counties"""
    plt.figure(figsize=(12, 8))

    score_columns = ['population_score', 'economic_score', 'mobility_score']
    score_data = most_affected_counties.loc[:, (slice(None), 'mean')][score_columns]

    sns.heatmap(score_data,
                annot=True,
                fmt='.1f',
                cmap='YlOrRd',
                yticklabels=[f"{c}, {s}" for (c, s) in score_data.index])

    plt.title('Vulnerability Scores for Most Affected Counties')
    return plt.gcf()

def plot_county_impact_trends(processed_dfs, top_n=10):
    """
    Plot day-wise impact scores for top affected counties

    Parameters:
    processed_dfs: List of processed DataFrames
    top_n: Number of top counties to plot
    """
    # Combine all data
    combined_data = pd.concat(processed_dfs)

    # Get top N counties by average impact score
    top_counties = combined_data.groupby(['County', 'State'])['total_impact_score'].mean()\
                               .nlargest(top_n).index

    # Create plot
    plt.figure(figsize=(15, 8))

    for county, state in top_counties:
        county_data = combined_data[
            (combined_data['County'] == county) &
            (combined_data['State'] == state)
        ]
        plt.plot(county_data['Day'],
                county_data['total_impact_score'],
                marker='o',
                label=f'{county}, {state}')

    plt.title(f'Impact Score Trends for Top {top_n} Most Affected Counties')
    plt.xlabel('Day')
    plt.ylabel('Impact Score')
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.grid(True)
    plt.tight_layout()

    return plt.gcf()

def plot_county_vulnerability_components(processed_dfs, top_n=10):
    """
    Plot day-wise vulnerability scores (population, economic, mobility) for top affected counties

    Parameters:
    processed_dfs: List of processed DataFrames
    top_n: Number of top counties to plot
    """
    # Combine all data
    combined_data = pd.concat(processed_dfs)

    # Get top N counties by average impact score
    top_counties = combined_data.groupby(['County', 'State'])['total_impact_score'].mean()\
                               .nlargest(top_n).index

    # Create subplots for each vulnerability component
    fig, axes = plt.subplots(3, 1, figsize=(15, 18))
    components = ['population_score', 'economic_score', 'mobility_score']
    titles = ['Population Vulnerability', 'Economic Vulnerability', 'Mobility Vulnerability']

    for idx, (component, title) in enumerate(zip(components, titles)):
        for county, state in top_counties:
            county_data = combined_data[
                (combined_data['County'] == county) &
                (combined_data['State'] == state)
            ]
            axes[idx].plot(county_data['Day'],
                         county_data[component],
                         marker='o',
                         label=f'{county}, {state}')

        axes[idx].set_title(f'{title} Score Trends')
        axes[idx].set_xlabel('Day')
        axes[idx].set_ylabel('Score')
        axes[idx].grid(True)
        axes[idx].legend(bbox_to_anchor=(1.05, 1), loc='upper left')

    plt.tight_layout()
    return fig

######
def create_county_impact_map(df, day, impact_threshold):
    """Creates a map visualizing county-level impact scores for a specific day."""

    # Filter for counties above the impact threshold
    high_impact_df = df[df['total_impact_score'] > impact_threshold]

    # Create figure
    fig = px.scatter_mapbox(high_impact_df,
                           lat='INTPTLAT',
                           lon='INTPTLON',
                           color='total_impact_score',  # Use 'total_impact_score' for color
                           size='total_impact_score',  # Use 'total_impact_score' for size
                           # color_continuous_scale='RdYlOrRd',  # Use a built-in colorscale
                           color_continuous_scale=px.colors.sequential.RdBu,  # Use a built-in colorscale
                           # Or, define a custom colorscale
                           # color_continuous_scale=[[0, 'blue'], [0.5, 'yellow'], [1, 'red']],
                           zoom=3,
                           center={'lat': 37.0902, 'lon': -95.7129},  # Center on the US
                           mapbox_style='carto-positron',
                           title=f"County-Level Impact on {day} (Impact Score > {impact_threshold})",
                           hover_name='County',
                           hover_data={'total_impact_score': True, 'State': True}
                           )

    fig.update_layout(margin={'r': 0, 't': 40, 'l': 0, 'b': 0})
    return fig


def plot_daily_impact_maps(processed_dfs, impact_threshold=60):
    """
    Create separate impact maps for each day
    Only shows counties with impact scores above the threshold

    Parameters:
    processed_dfs: List of processed DataFrames
    impact_threshold: Minimum impact score to display (default: 60)
    """
    maps = []

    # Track daily statistics
    daily_stats = []

    for df in processed_dfs:
        day = df['Day'].iloc[0]

        # Calculate daily statistics
        total_counties = len(df)
        affected_counties = len(df[df['total_impact_score'] >= impact_threshold])
        affected_percentage = (affected_counties / total_counties) * 100

        daily_stats.append({
            'Day': day,
            'Total Counties': total_counties,
            'Affected Counties': affected_counties,
            'Affected Percentage': affected_percentage
        })

        # Create map
        fig = create_county_impact_map(df, day, impact_threshold)
        maps.append(fig)

    # Create summary DataFrame
    stats_df = pd.DataFrame(daily_stats)
    print("\nDaily Impact Statistics:")
    print(stats_df.to_string(index=False))

    return maps
```

```python
# Section 6: Generate Final Report
def generate_comprehensive_report(processed_dfs, combined_results):
    """Generate final analysis report"""
    most_affected_counties, most_affected_states = identify_most_affected_areas(combined_results)

    report = {
        'overall_statistics': {
            'total_counties_analyzed': len(combined_results['County'].unique()),
            'total_states_analyzed': len(combined_results['State'].unique()),
            'average_impact_score': combined_results['total_impact_score'].mean(),
            'maximum_impact_score': combined_results['total_impact_score'].max()
        },
        'most_affected_counties': most_affected_counties,
        'most_affected_states': most_affected_states,
        'daily_trends': {
            df['Day'].iloc[0]: {
                'avg_impact_score': df['total_impact_score'].mean(),
                'max_impact_score': df['total_impact_score'].max(),
                'most_affected_county': df.loc[df['total_impact_score'].idxmax(), 'County']
            } for df in processed_dfs
        }
    }

    return report
```
